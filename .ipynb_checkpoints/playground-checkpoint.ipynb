```python
import shutil
import kagglehub
import os
import nltk
from nltk.stem import WordNetLemmatizer
import nltk
from nltk.corpus import stopwords
import ssl
import json
import pandas as pd

try:
    _create_unverified_https_context = ssl._create_unverified_context
except AttributeError:
    pass
else:
    ssl._create_default_https_context = _create_unverified_https_context
nltk.download('stopwords')
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()


## To lemmatize a sentence
def lemmatize_words(sentence):
    punctuation = set('''!()-[]{};:'"\\,<>./?@#$%^&*_~''')
    words = []
    current_word = []
    
    for char in sentence:
        if char.isspace() or char in punctuation:
            if current_word:
                words.append(''.join(current_word))
                current_word = []
        else:
            current_word.append(char)
    
    # Add the last word if exists
    if current_word:
        word = ''.join(current_word)
        word = lemmatizer.lemmatize(word, pos='v')
        words.append(word)
    
    for w in range(len(words)):
        words[w] = lemmatizer.lemmatize(words[w].lower(), pos='v')
    
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]

    return words

## To count the occurrence of words in a sentence
def count_words(array_of_sentences, count_dict=None):
    if count_dict is None:
        count_dict = {}
    for sentence in array_of_sentences:
        for word in sentence:
            count_keys = count_dict.keys()
            if word in count_keys:
                count_dict[word] += 1
            else:
                count_dict[word] = 1
    return count_dict

## To read an json
def read_json(path):
    with open(path, 'r') as f:
        data = json.load(f)
    return data

## To dump into json
def dump_json(data, path):
    with open(path, 'w+') as f:
        json.dump(data, f, indent=4)
```

```python

def union_count_dicts(count_dict1, count_dict2, ignore_words = []):
    """
    Combine two count dictionaries by adding counts for common keys.
    
    Args:
        count_dict1: First dictionary with counts
        count_dict2: Second dictionary with counts
    
    Returns:
        A new dictionary with the sum of counts for each key
    """
    result = count_dict1.copy()  # Start with all items from first dict
    
    for key, value in count_dict2.items():
        if key in ignore_words:
            continue
        if key in result:
            result[key] += value  # Add counts if key exists in both
        else:
            result[key] = value  # Add new key-value pair
    return result

def get_top_n_largest(count_dict, n=[0,15]):
    """
    Get the top n keys with the largest values from a count dictionary.
    
    Args:
        count_dict: Dictionary with keys and their counts
        n: Number of top items to return (default: 15)
    
    Returns:
        Dictionary with top n key-value pairs sorted by value descending
    """
    if not count_dict:
        return {}
    
    # Use heapq.nlargest for better performance with large dictionaries
    import heapq
    top_items = heapq.nlargest(n[1], count_dict.items(), key=lambda x: x[1])
    return dict(top_items[n[0]:])
```

```python
oscar_good_count_dict = {}
for movie in oscar_good_dataset:
    oscar_good_count_dict = union_count_dicts(movie['critic_good_reviews'], oscar_good_count_dict)

oscar_bad_count_dict = {}
for movie in oscar_bad_dataset:
    oscar_bad_count_dict = union_count_dicts(movie['critic_bad_reviews'], oscar_bad_count_dict)

laymen_good_count_dict = {}
for movie in laymen_good_dataset:
    laymen_good_count_dict = union_count_dicts(movie['audience_reviews'], laymen_good_count_dict)

laymen_bad_count_dict = {}
for movie in laymen_bad_dataset:
    laymen_bad_count_dict = union_count_dicts(movie['audience_reviews'], laymen_bad_count_dict)

```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, IntSlider
from IPython.display import clear_output

def pack_circles(radii, padding=0.1, max_iterations=1000):
    """Improved circle packing algorithm"""
    n = len(radii)
    positions = [(0, 0)]
    
    for i in range(1, n):
        best_pos = None
        min_overlap = float('inf')
        
        for j in range(i):
            for angle in np.linspace(0, 2*np.pi, 60):
                x = positions[j][0] + (radii[i] + radii[j] + padding) * np.cos(angle)
                y = positions[j][1] + (radii[i] + radii[j] + padding) * np.sin(angle)
                
                total_overlap = 0
                for k, (xk, yk) in enumerate(positions[:i]):
                    dist = np.sqrt((x - xk)**2 + (y - yk)**2)
                    overlap = max(0, (radii[i] + radii[k] + padding) - dist)
                    total_overlap += overlap
                
                if total_overlap < min_overlap - 1e-6:
                    min_overlap = total_overlap
                    best_pos = (x, y)
                    
                if min_overlap == 0:
                    break
            if min_overlap == 0:
                break
        
        if best_pos:
            positions.append(best_pos)
        else:
            angle = (i * 137.5) * np.pi / 180
            radius_sum = radii[i] + max(radii[:i])
            x = radius_sum * np.cos(angle)
            y = radius_sum * np.sin(angle)
            positions.append((x, y))
    
    return positions

def plot_graph(graph_type):
    # Clear previous output
    clear_output(wait=True)
    
    # Create new figure
    fig, ax = plt.subplots(figsize=(12, 10))
    
    # Get data (assuming oscar_bad_count_dict exists)
    values = []
    labels = []
    index_value = graph_type
    for key, value in oscar_bad_count_dict[index_value].items():
        values.append(value)
        labels.append(key)

    # Calculate radii
    total_area = sum(values)
    max_radius = 2
    radii = [np.sqrt(value / total_area) * max_radius * 2 for value in values]
    
    # Pack circles
    positions = pack_circles(radii, padding=0.05)

    ax.axis('off')
    
    # Find bounds to center the visualization
    xs, ys = zip(*positions)
    x_min, x_max = min(xs) - max(radii), max(xs) + max(radii)
    y_min, y_max = min(ys) - max(radii), max(ys) + max(radii)
    padding_val = 1
    ax.set_xlim(x_min - padding_val, x_max + padding_val)
    ax.set_ylim(y_min - padding_val, y_max + padding_val)
    
    colors = plt.cm.Set3(np.linspace(0, 1, len(values)))
    
    for (x, y), radius, value, label, color in zip(positions, radii, values, labels, colors):
        circle = plt.Circle((x, y), radius, color=color, alpha=0.7, ec='black', linewidth=1.5)
        ax.add_patch(circle)
        fontsize = max(8, min(12, radius * 4))
        ax.text(x, y, f'{label}\n{value}', ha='center', va='center', 
                fontsize=fontsize, fontweight='bold')
    
    ax.set_title(f'Most used words in audience reviews of Oscar Movies - From Year {1980+graph_type*5}-{1980+graph_type*5+4}', fontsize=16, pad=20)
    ax.set_aspect('equal')
    plt.tight_layout()
    plt.show()

# Create interactive slider
interact(plot_graph, graph_type=IntSlider(min=0, max=7, step=1, description='Graph:'))
```

```python
def counts_to_probs(counts):
    """Convert a dictionary of counts to a dictionary of probabilities."""
    total = sum(counts.values())
    # Avoid division by zero if total is 0
    if total == 0:
        return {key: 0 for key in counts}
    return {key: value / total for key, value in counts.items()}

def get_top_n_largest(count_dict, n=[0,15]):
    """
    Get the top n keys with the largest values from a count dictionary.
    
    Args:
        count_dict: Dictionary with keys and their counts
        n: Number of top items to return (default: 15)
    
    Returns:
        Dictionary with top n key-value pairs sorted by value descending
    """
    if not count_dict:
        return {}
    
    # Use heapq.nlargest for better performance with large dictionaries
    import heapq
    top_items = heapq.nlargest(n[1], count_dict.items(), key=lambda x: x[1])
    top_items = dict(top_items[n[0]:])
    top_items = {key: round(value, 2) for key, value in top_items.items() if value > 0.0001}
    return top_items
```

```python
oscar_good_keywords = {}
for movie in oscar_good_dataset:
    sentence = movie['keywords']
    sentence = count_words([sentence])
    oscar_good_keywords = union_count_dicts(sentence, oscar_good_keywords)

oscar_bad_keywords = {}
for movie in oscar_bad_dataset:
    sentence = movie['keywords']
    sentence = count_words([sentence])
    oscar_bad_keywords = union_count_dicts(sentence, oscar_bad_keywords)

laymen_good_keywords = {}
for movie in laymen_good_dataset:
    sentence = movie['keywords']
    sentence = count_words([sentence])
    laymen_good_keywords = union_count_dicts(sentence, laymen_good_keywords)

laymen_bad_keywords = {}
for movie in laymen_bad_dataset:
    sentence = movie['keywords']
    sentence = count_words([sentence])
    laymen_bad_keywords = union_count_dicts(sentence, laymen_bad_keywords)

```

<h1>What genre of Oscar movies are Oscar-baity actors mostly in?</h1>

```python
def is_actor_in_this_movie(cast_dict, list_of_actors):
    for x in cast_dict:
        if x['name'] in list_of_actors:
            return True
    return False
```

```python
genre_count = {}
genre_bad_count = {}
genre_good_count = {}
all_count = 0
bad_count = 0
good_count = 0
list_of_oscar_bait_actors = list(get_top_n_largest(oscar_bait_actor_count, n=[0,15]).keys())
i = 0
size = len(oscar_dataset)
for movie in oscar_dataset:
    i+=1
    if is_actor_in_this_movie(movie['cast'], list_of_oscar_bait_actors):
        all_count += 1
    # if True:
        good_count_temp, bad_count_temp = False, False
        for genre in movie['genres']:
            if movie['tomatometer_rating'] is not None:
                if movie['tomatometer_rating'] <= 60:
                    append_count(genre_bad_count, genre)
                    bad_count_temp = True
                else:
                    append_count(genre_good_count, genre)
                    good_count_temp = True
            append_count(genre_count, genre)
        bad_count = bad_count + 1 if bad_count_temp else bad_count
        good_count = good_count + 1 if good_count_temp else good_count

    if i % 100 == 0:
        print(f"\r{i}/{size} movies processed. ({i/size*100:.1f}%)", end="", flush=True)
```

```python
def divide_count_dict(count_dict, value):
    return {key: round(val / value, 3) for key, val in count_dict.items()}
```

```python
genre_count = get_top_n_largest(genre_count,n=[0,4])
genre_bad_count = get_top_n_largest(genre_bad_count,n=[0,4])
genre_good_count = get_top_n_largest(genre_good_count,n=[0,4])
```

```python
divide_count_dict(genre_count, all_count), divide_count_dict(genre_bad_count, bad_count), divide_count_dict(genre_good_count, good_count)
```

<h1>Most common directors in Oscar movies</h1>

```python
def append_count(dic, key):
    if key in list(dic.keys()):
        dic[key] += 1
    else:
        dic[key] = 1
    return dic
oscar_director_count = {}
oscar_good_director_count = {}
oscar_bad_director_count = {}
i = 0
size = len(oscar_dataset)
for movie in oscar_dataset:
    i += 1
    if movie is None:
        continue
    if movie['tomatometer_rating'] is None:
        continue
    for crew_member in movie['crew']:
        if crew_member['job'] != "Director":
            continue
        append_count(oscar_director_count, crew_member['name'])
        if movie['tomatometer_rating'] > 60:
            append_count(oscar_good_director_count, crew_member['name'])
        else:
            append_count(oscar_bad_director_count, crew_member['name'])
            
    if i % 100 == 0:
        print(f"\r{i}/{size} movies processed. ({i/size*100:.1f}%)", end="", flush=True)
```
