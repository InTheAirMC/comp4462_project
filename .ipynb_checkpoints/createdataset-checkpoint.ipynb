<h2>Create JSON data sets of the corresponding downloaded kaggle datasets</h2>

```python
import nltk
from nltk.stem import WordNetLemmatizer
import nltk
from nltk.corpus import stopwords
import ssl
import json
import pandas as pd
try:
    _create_unverified_https_context = ssl._create_unverified_context
except AttributeError:
    pass
else:
    ssl._create_default_https_context = _create_unverified_https_context
nltk.download('stopwords')
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()


## To lemmatize a sentence
def lemmatize_words(sentence):
    punctuation = set('''!()-[]{};:'"\\,<>./?@#$%^&*_~''')
    words = []
    current_word = []
    
    for char in sentence:
        if char.isspace() or char in punctuation:
            if current_word:
                words.append(''.join(current_word))
                current_word = []
        else:
            current_word.append(char)
    
    # Add the last word if exists
    if current_word:
        word = ''.join(current_word)
        word = lemmatizer.lemmatize(word, pos='v')
        words.append(word)
    
    for w in range(len(words)):
        words[w] = lemmatizer.lemmatize(words[w].lower(), pos='v')
    
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]

    return words

## To count the occurrence of words in a sentence
def count_words(array_of_sentences, count_dict=None):
    if count_dict is None:
        count_dict = {}
    for sentence in array_of_sentences:
        for word in sentence:
            count_keys = count_dict.keys()
            if word in count_keys:
                count_dict[word] += 1
            else:
                count_dict[word] = 1
    return count_dict

## To read an json
def read_json(path):
    with open(path, 'r') as f:
        data = json.load(f)
    return data

## To dump into json
def dump_json(data, path):
    with open(path, 'w+') as f:
        json.dump(data, f, indent=4)
```

<h1> Convert the Critic Reviews to JSON</h1>

```python
import pandas as pd
import os
import ast

class CriticReview:
    def __init__(self):
        self.critic_reviews = pd.read_csv(os.getcwd() + "/datasets/rottentomatoes-review-dataset/rotten_tomatoes_critic_reviews.csv")
        self.movies = pd.read_csv(os.getcwd() + "/datasets/rottentomatoes-review-dataset/rotten_tomatoes_movies.csv")
    
    def get_all_movie_id(self):
        return self.movies['rotten_tomatoes_link'].tolist()
        
    def output_dataset(self):
        i = 0
        movie_ids = self.get_all_movie_id()
        output_dataset = {}
        total_movies = len(movie_ids)
        
        for movie_id in movie_ids:
            i += 1
            output_dataset[movie_id] = {}
            
            # Get movie data once for efficiency
            movie_data = self.movies[self.movies['rotten_tomatoes_link'] == movie_id]
            
            # Input movie title (with error handling)
            if not movie_data.empty:
                output_dataset[movie_id]['movie_title'] = movie_data['movie_title'].values[0] if 'movie_title' in movie_data else 'Unknown'
                output_dataset[movie_id]['release_date'] = movie_data['original_release_date'].values[0] if 'original_release_date' in movie_data else ''
                output_dataset[movie_id]['tomatometer_status'] = movie_data['tomatometer_status'].values[0] if 'tomatometer_status' in movie_data else ''
                output_dataset[movie_id]['tomatometer_rating'] = movie_data['tomatometer_rating'].values[0] if 'tomatometer_rating' in movie_data else 0
                output_dataset[movie_id]['audience_status'] = movie_data['audience_status'].values[0] if 'audience_status' in movie_data else ''
                output_dataset[movie_id]['audience_rating'] = movie_data['audience_rating'].values[0] if 'audience_rating' in movie_data else 0
                output_dataset[movie_id]['movie_info'] = movie_data['movie_info'].values[0] if 'movie_info' in movie_data else ''
                output_dataset[movie_id]['critics_consensus'] = movie_data['critics_consensus'].values[0] if 'critics_consensus' in movie_data else ''
            
            # Get critic reviews for this movie
            critic_reviews_data = self.critic_reviews[self.critic_reviews['rotten_tomatoes_link'] == movie_id]
            
            list_of_critic_reviews = critic_reviews_data['review_content'].dropna().tolist()
            
            # Fresh reviews (positive)
            list_of_fresh_reviews = self.critic_reviews[
                (self.critic_reviews['rotten_tomatoes_link'] == movie_id) & 
                (self.critic_reviews['review_type'].astype(str).str.lower() == 'fresh')
            ]['review_content'].dropna().tolist()
            
            # Rotten reviews (negative)
            list_of_rotten_reviews = self.critic_reviews[
                (self.critic_reviews['rotten_tomatoes_link'] == movie_id) & 
                (self.critic_reviews['review_type'].astype(str).str.lower() == 'rotten')
            ]['review_content'].dropna().tolist()
            
            # Process reviews
            array_of_sentences = []
            array_of_fresh_sentences = []
            array_of_rotten_sentences = []
            
            # Process fresh reviews
            for review in list_of_fresh_reviews:
                review = lemmatize_words(str(review))
                array_of_fresh_sentences.append(review)
                array_of_sentences.append(review)
            output_dataset[movie_id]['critic_good_reviews'] = count_words(array_of_fresh_sentences)
            
            # Process rotten reviews
            for review in list_of_rotten_reviews:
                review = lemmatize_words(str(review))
                array_of_rotten_sentences.append(review)
                array_of_sentences.append(review)
            output_dataset[movie_id]['critic_bad_reviews'] = count_words(array_of_rotten_sentences)
            
            # Process all reviews
            output_dataset[movie_id]['critic_reviews'] = count_words(array_of_sentences)
            
            # Add statistics
            output_dataset[movie_id]['review_stats'] = {
                'total_reviews': len(list_of_critic_reviews),
                'fresh_count': len(list_of_fresh_reviews),
                'rotten_count': len(list_of_rotten_reviews)
            }
            
            # Progress indicator
            print(f"\r{i}/{total_movies} movies processed. ({i/total_movies*100:.1f}%)", end="", flush=True)
            
            # Save checkpoint every 10 movies (you had 10, not 100)
            if i % 1000 == 0:
                dump_json(output_dataset, os.getcwd() + "/critic_reviews.json")
        
        # Final save
        dump_json(output_dataset, os.getcwd() + "/critic_reviews.json")
        print("\n✅ Done!")
        return output_dataset
```

<h1>Convert Oscars Dataset to JSON</h1>

```python
class OscarsDataset:
    def __init__(self):
        self.oscar_award = pd.read_csv(os.getcwd() + "/datasets/the-oscar-award/the_oscar_award.csv")
    
    def get_all_movie_names(self):
        return list(set(self.oscar_award['film'].tolist()))
    
    def output_dataset(self):
        movie_names = self.get_all_movie_names()
        output_dataset = {}
        i = 0
        for movie_name in movie_names:
            i+=1
            output_dataset[movie_name] = {}
            output_dataset[movie_name]['oscars_awards'] = {"count":0,"categories":[], "wins":0}
            for index, oscar in self.oscar_award[self.oscar_award['film'] == movie_name].iterrows():
                canon_category = oscar['canon_category']
                winner = oscar['winner']
                output_dataset[movie_name]['year'] = oscar['year_film']
                if winner == 'True' or winner == True:
                    output_dataset[movie_name]['oscars_awards']['wins'] += 1
                output_dataset[movie_name]['oscars_awards']['categories'].append({"award_name":canon_category,"winner":winner})
                output_dataset[movie_name]['oscars_awards']['count'] += 1
            if i % 100 == 0:
                dump_json(output_dataset, os.getcwd() + "/oscar_awards.json")
            print(f"\r{i}/{len(movie_names)} movies processed.", end="",flush=True)

        dump_json(output_dataset, os.getcwd() + "/oscar_awards.json")
        print("\nDone.")
        return output_dataset
```

```python
import pandas as pd
import os
import ast
import json

class TheMoviesDataset:
    def __init__(self):
        self.movies_metadata = pd.read_csv(os.getcwd() + "/datasets/the-movies-dataset/movies_metadata.csv")
        self.keywords = pd.read_csv(os.getcwd() + "/datasets/the-movies-dataset/keywords.csv")
        self.credits = pd.read_csv(os.getcwd() + "/datasets/the-movies-dataset/credits.csv")
        
    def output_all_movie_ids(self):
        """Return unique movie IDs"""
        return self.movies_metadata['id'].drop_duplicates().tolist()
    
    def safe_literal_eval(self, value):
        """Safely evaluate string to Python object"""
        try:
            if pd.isna(value) or value == '' or value == '[]':
                return []
            return ast.literal_eval(value)
        except (ValueError, SyntaxError, TypeError) as e:
            print(f"Warning: Could not parse: {value[:50]}... Error: {e}")
            return []
    
    def output_dataset(self):
        movie_ids = self.output_all_movie_ids()
        output_dataset = {}
        total = len(movie_ids)
        
        for idx, movieid in enumerate(movie_ids, 1):
            try:
                # Convert to string for consistent comparison
                movieid_str = str(movieid)
                
                # Get movie metadata
                movie_row = self.movies_metadata[self.movies_metadata['id'].astype(str) == movieid_str]
                
                if movie_row.empty:
                    print(f"\nWarning: Movie ID {movieid} not found in metadata")
                    continue
                
                movie = movie_row.iloc[0]
                
                # Initialize movie entry
                output_dataset[movieid_str] = {}
                
                # Basic info
                output_dataset[movieid_str]["budget"] = movie.get('budget', 0)
                output_dataset[movieid_str]["name"] = movie.get('title', 'Unknown')
                output_dataset[movieid_str]["revenue"] = movie.get('revenue', 0)
                output_dataset[movieid_str]["runtime"] = movie.get('runtime', 0)
                output_dataset[movieid_str]["release_date"] = movie.get('release_date', '')
                
                # Process genres
                genres = self.safe_literal_eval(movie.get('genres', '[]'))
                output_dataset[movieid_str]["genres"] = []
                for x in genres:
                    if isinstance(x, dict) and 'name' in x:
                        output_dataset[movieid_str]["genres"].append(x['name'])
                
                # Process spoken languages
                spoken_languages = self.safe_literal_eval(movie.get('spoken_languages', '[]'))
                output_dataset[movieid_str]["spoken_languages"] = []
                for x in spoken_languages:
                    if isinstance(x, dict) and 'name' in x:
                        output_dataset[movieid_str]["spoken_languages"].append(x['name'])
                
                # Process keywords
                keywords_data = self.keywords[self.keywords['id'].astype(str) == movieid_str]
                output_dataset[movieid_str]["keywords"] = []
                if not keywords_data.empty:
                    keyword_row = keywords_data.iloc[0]
                    if 'keywords' in keyword_row and pd.notna(keyword_row['keywords']):
                        keywords_list = self.safe_literal_eval(keyword_row['keywords'])
                        output_dataset[movieid_str]["keywords"] = [
                            k['name'] for k in keywords_list 
                            if isinstance(k, dict) and 'name' in k
                        ]
                
                # Process credits (cast)
                credits_data = self.credits[self.credits['id'].astype(str) == movieid_str]
                output_dataset[movieid_str]["cast"] = []
                if not credits_data.empty:
                    credits_row = credits_data.iloc[0]
                    if 'cast' in credits_row and pd.notna(credits_row['cast']):
                        cast_list = self.safe_literal_eval(credits_row['cast'])
                        output_dataset[movieid_str]["cast"] = [
                            {'name': k.get('name', 'Unknown'), 'gender': k.get('gender', 0)} 
                            for k in cast_list if isinstance(k, dict)
                        ]
                
                # Optional: Add crew information
                output_dataset[movieid_str]["crew"] = []
                if not credits_data.empty and 'crew' in credits_row:
                    crew_list = self.safe_literal_eval(credits_row['crew'])
                    output_dataset[movieid_str]["crew"] = [
                        {'name': k.get('name', 'Unknown'), 'job': k.get('job', '')}
                        for k in crew_list if isinstance(k, dict)
                    ]
                
            except Exception as e:
                print(f"\nError processing movie ID {movieid}: {e}")
                continue
            
            # Progress indicator
            print(f"\rProcessing movies: {idx}/{total} ({idx/total*100:.1f}%)", 
                  end="", flush=True)
            if idx % 100 == 0:
                dump_json(output_dataset, os.getcwd() + "/the_movies_dataset.json")

        dump_json(output_dataset, os.getcwd() + "/the_movies_dataset.json")
        print("\n✅ Dataset creation complete!")
        return output_dataset
    
```

<h1>Convert the IMDB reviews to JSON</h1>

```python
import re

def extract_movie_info(text):
    """Extract movie title, year, and range if present"""
    
    # Pattern captures: title, year, optional range, and optional extra text
    pattern = r'^(.+?)\s+\((\d{4})(?:[\u2013\-–]([^)]+))?\)(.*?)$'
    
    match = re.match(pattern, text.strip())
    if match:
        title = match.group(1).strip()
        year = match.group(2)
        range_info = match.group(3).strip() if match.group(3) else None
        extra_text = match.group(4).strip() if match.group(4) else None
        
        return {
            'title': title,
            'start_year': year,
            'is_movie': not bool(range_info or extra_text),  # True if no range and no extra text
            'range': range_info,
            'extra': extra_text
        }
    return None

# Test
result = extract_movie_info('Joy to the world (2020) Season 1, Episode 2')
print(result)
# Output: {'title': 'Joy to the world', 'start_year': '2020', 'is_movie': False, 'range': None, 'extra': 'Season 1, Episode 2'}
```

```python
class IMDBReviews:
    def __init__(self):
        print(f"\rInitialising 1/6",end="", flush=True)
        self.part_01 = read_json(os.getcwd() + "/datasets/imdb-review-dataset/part-01.json")
        # self.part_02 = self.part_03 = self.part_04 = self.part_05 = self.part_06 = []
        print(f"\rInitialising 2/6",end="", flush=True)
        self.part_02 = read_json(os.getcwd() + "/datasets/imdb-review-dataset/part-02.json")
        print(f"\rInitialising 3/6",end="", flush=True)
        self.part_03 = read_json(os.getcwd() + "/datasets/imdb-review-dataset/part-03.json")
        print(f"\rInitialising 4/6",end="", flush=True)
        self.part_04 = read_json(os.getcwd() + "/datasets/imdb-review-dataset/part-04.json")
        print(f"\rInitialising 5/6",end="", flush=True)
        self.part_05 = read_json(os.getcwd() + "/datasets/imdb-review-dataset/part-05.json")
        print(f"\rInitialising 6/6",end="", flush=True)
        self.part_06 = read_json(os.getcwd() + "/datasets/imdb-review-dataset/part-06.json")
        print("\nInitialised")

    def output_dataset(self, maxtime = 10000):
        output_dataset = {}
        i = 0
        fail = 0
        length = len(self.part_01) + len(self.part_02) + len(self.part_03) + len(self.part_04) + len(self.part_05) + len(self.part_06) 
        for dataset in [self.part_01, self.part_02, self.part_03, self.part_04, self.part_05, self.part_06]:
            for review in dataset:
                i += 1
                if i >= maxtime:
                    break
                try:
                    movie_name = review['movie']
                    # print(movie_name)
                    movie_info = extract_movie_info(movie_name)
                    # print(movie_info)
                    if movie_info['is_movie'] is None or movie_info['is_movie'] == False: #Skip if it is a series
                        continue
                    if not movie_name in list(output_dataset.keys()):
                        output_dataset[movie_name] = {}
                        output_dataset[movie_name]['title'] = movie_info['title']
                        output_dataset[movie_name]['year'] = movie_info['start_year']
                        output_dataset[movie_name]['rating'] = {'score':0, 'count':0}
                        output_dataset[movie_name]['reviews'] = {}
                    review_detail = lemmatize_words(review['review_detail'])
                    output_dataset[movie_name]['reviews'] = count_words([review_detail], count_dict=output_dataset[movie_name]['reviews'])
                    
                    if isinstance(review['rating'], str):
                        score = output_dataset[movie_name]['rating']['score'] 
                        count = output_dataset[movie_name]['rating']['count']
                        score = score * count + int(review['rating'])
                        count += 1
                        score /= count
                        output_dataset[movie_name]['rating']['score'] = score
                        output_dataset[movie_name]['rating']['count'] = count
                        
                    
                    if i % 100 == 0:
                        dump_json(output_dataset, os.getcwd() + "/imdb_dataset.json")
                except:
                    fail +=1
                print(f"\rProcessing movies: {i}/{length} ({i/length*100:.1f}%)|||Failed: {fail}", end="", flush=True)
        print("\nDone")
        dump_json(output_dataset, os.getcwd() + "/imdb_dataset.json")
```

<h1>Merge all datasets into movies.json</h1>

```python
class MoviesDataset:
    def __init__(self):
        self.movies = read_json('movies.json')
    def __del__(self):
        dump_json(self.movies, 'movies.json')
    def initalise(self):
        movies_dataset = read_json('the_movies_dataset.json')
        oscars_dataset = read_json('oscar_awards.json')
        critic_dataset = read_json('critic_reviews.json')
        imdb_dataset = read_json('imdb_dataset.json')

        self.movies = []
        # We use the movies dataset's movie name as our "anchor"
        i = 0
        fail = 0
        total = len(list(movies_dataset.keys()))
        for key, values in movies_dataset.items():
            i += 1
            try:
                this_movie = {}
                name = values['name']
                this_movie['movie_name'] = name
                this_movie['budget'] = values['budget']
                this_movie['revenue'] = values['revenue']
                this_movie['runtime'] = values['runtime']
                year = int(values['release_date'].split('-')[0])
                this_movie['year'] = year
                this_movie['genres'] = values['genres']
                this_movie['spoken_languages'] = values['spoken_languages']
                this_movie['keywords'] = values['keywords']
                this_movie['cast'] = values['cast']
                this_movie['crew'] = values['crew']
                #Input Oscar Details
                this_movie['oscars'] = {"count":0, "categories":[], "wins":0}
                if name in list(oscars_dataset.keys()): #If the movie is an oscar movie
                    this_movie['oscars']['count'] = oscars_dataset[name]['oscars_awards']['count']
                    this_movie['oscars']['wins'] = oscars_dataset[name]['oscars_awards']['wins']
                    this_movie['oscars']['categories'] = oscars_dataset[name]['oscars_awards']['categories']
            
                #Input Critics Dataset
                this_movie['tomatometer_status'] = None
                this_movie['tomatometer_rating'] = None
                this_movie['audience_status'] = None
                this_movie['audience_rating'] = None
                this_movie['movie_info'] = None
                this_movie['critics_consensus'] = None
                this_movie['critic_good_reviews'] = None
                this_movie['critic_bad_reviews'] = None
                this_movie['review_status'] = {"total_reviews": 0,"fresh_count": 0,"rotten_count": 0}
    
                for critic_id, critic_values in critic_dataset.items():
                    if critic_values['movie_title'] != name:
                        continue
                    else:
                        this_movie['tomatometer_status'] = critic_values['tomatometer_status']
                        this_movie['tomatometer_rating'] = critic_values['tomatometer_rating']
                        this_movie['audience_status'] = critic_values['audience_status']
                        this_movie['audience_rating'] = critic_values['audience_rating']
                        this_movie['movie_info'] = critic_values['movie_info']
                        this_movie['critics_consensus'] = critic_values['critics_consensus']
                        this_movie['critic_good_reviews'] = critic_values['critic_good_reviews']
                        this_movie['critic_bad_reviews'] = critic_values['critic_bad_reviews']
                        this_movie['review_stats'] = critic_values['review_stats']
                        break
    
                #Input IMDB Dataset
                this_movie['audience_reviews'] = {}
                for imdb_id, imdb_values in imdb_dataset.items():
                    if imdb_values['title'] != name:
                        continue
                    else:
                        this_movie['audience_reviews'] = imdb_values['reviews']
                        break
                self.movies.append(this_movie)
            except:
                fail +=1
            
            print(f"\rProcessing {str(i)}/{str(total)} Movies || Fail: {str(fail)} || {name} ",end="", flush=True)
            
    def dump_to_json(self):
        dump_json(self.movies, 'movies.json')
```

```python
import json
import pandas as pd

# Load the JSON file
with open('movies.json', 'r') as file:
    data = json.load(file)

# Function to flatten nested structures
def flatten_movie(movie):
    flattened = {"oscarsYesNo":value['oscars']['count']}
    
    for key, value in movie.items():
        if isinstance(value, list):
            # Convert lists to JSON strings or join them
            if key == 'genres':
                flattened[key] = ', '.join(value)
            elif key == 'spoken_languages':
                flattened[key] = ', '.join(value)
            elif key == 'keywords':
                flattened[key] = ', '.join(value)
            elif key in ['cast', 'crew']:
                # Convert list of dicts to JSON string
                flattened[key] = json.dumps(value)
            else:
                flattened[key] = json.dumps(value)
                
        elif isinstance(value, dict):
            # Flatten nested dictionaries
            for sub_key, sub_value in value.items():
                flattened[f"{key}_{sub_key}"] = sub_value
        else:
            flattened[key] = value
    flattened["OscarsYesno"] = value['oscars']['count']

    return flattened

# Flatten each movie
i = 0
flattened_data = []
for movie in data:
    i +=1 
    # if i >= 10000:
    #     break
    movie['OscarsYesNo'] = movie['oscars']['count']
    movie['OscarsYesNoWins'] = movie['oscars']['wins']
    flattened_data.append(movie)

# Convert to DataFrame and save as CSV
df = pd.DataFrame(flattened_data)
df.to_csv('movies.csv', index=False)

print("CSV file created successfully!")
```
